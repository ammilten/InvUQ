# Problem Setup


The example we use is a simple 1D seismic problem. We assume a homogeneous background material with a single layer of some thickness at some depth. We can vary the wavelet frequency, too, but in this dataset we will assume it is known. Here is a simple diagram showing the setup.

![title](SeismicExampleSetup.png)

# Step 1. Import the Data

We have already simulated the data for this problem and have saved it in a pickle file. Here we load the data, add some noise, and visualize one data example.

```python
data = pkl.load(open('seismic_data.pkl','rb'))

ztrue = data["z"][0,:]
ptrue = data["p"][0,:]

zprior = data["z"][1:,:]
pprior = data["p"][1:,:]

znoise = multivariate_normal.rvs(np.zeros(zprior.shape[1]), data["errcov"], zprior.shape[0])
```

```python
plt.plot(ztrue+znoise[0,:], data["tvec"])
plt.gca().invert_yaxis()
plt.ylabel('Two-Way Travel Time (s)')
plt.xlabel('Amplitude')
```

# Step 2. Prior Falsification with PCA

We use this custom PosteriorFalsification class to make eigenvalue decomposition and analysis simple. In this example the data and prior are both from the same Monte Carlo simulation, so the observed ("true") data will definitely fall within the prior.

The PosteriorFalsification class also comes with functions to plot score cross plots and a scree plot.

```python
PF = PriorFalsification(zprior+znoise, ztrue[np.newaxis,:])
```

```python
PF.plot_scores()
```

```python
PF.plot_scree()
```

# Step 3. Training the ML Inverse Model

Now we can train our inverse models. We use Random Forest Regressions for each individual parameter, but it could be anything. 

After training the models we package them into a neat function that maps $p^*=f^{-1}(z)$ where $p^*$ is the predicted parameter vector, $f^{-1}$ is the machine learning inverse model, and $z$ is the data vector.

```python
def train_RF(D, inputs):
    '''
    Inputs:
        D: [SxD] matrix of training data
        inputs: [SxP] matrix of training parameters

    Outputs:
        models: [Px1] list of random forest regressors
    '''
    nparams = inputs.shape[1]
    models = [None]*nparams
    for i in range(nparams):
        print('Training '+str(i+1)+' of '+str(nparams))
        models[i] = RandomForestRegressor().fit(D, inputs[:,i])

    def MLinv(D):
        params = np.zeros((D.shape[0], len(models)))
        for i in range(len(models)):
            params[:,i] = models[i].predict(D)
        return params

    return MLinv
```

We will use the first 7500 samples as our training set and the rest will be the test set

```python
itrain = 7500

ztrain = zprior[:itrain,:]+znoise[:itrain,:]
MLinv = train_RF(ztrain, pprior[:itrain,:])
```

Now we can use our ML inverse model to generate predictions. Let's take a look at the predictions.

```python
p_pred_prior = MLinv(zprior+znoise)

utils.plot_predictions(p_pred_prior, pprior, ntrain=itrain, varnames=data["varnames"], units=data["units"])
```

# Step 4. Posterior Sampling

Now that we have our test sets of true ($p_{test}$) and predicted ($p^*_{test}$) parameters, we can perform posterior sampling conditioned on the observed data $z_{obs}$. First, we sample $N_{obs}$ realizations of $z_{obs}$ from $\textrm{Pr}(z_{obs})$, which in this case is a multivariate Gaussian.

```python
Nobs = 10000
zobs = multivariate_normal.rvs(ztrue, data["errcov"], Nobs)
```

Next, we generate our predicted parameter set based on the observed data, $p^*_{obs}=f^{-1}(z_{obs})$. Through a prediction plot we can see how the ML inverse model and data uncertainty are related.

```python
p_pred_obs = MLinv(zobs)

utils.plot_predictions(p_pred_prior[itrain:,:], pprior[itrain:,:], p_pred_obs=p_pred_obs, p_true_obs=ptrue, varnames=data["varnames"], units=data["units"])
```

Now that we have our three parameter sets, $p_{test}$, $p^*_{test}$, and $p_{obs}^*$, we can use the recently-developed empirical posterior sampling method to generate $p_{obs}$. Right now this only generates uncorrelated marginal posteriors, but joint posterior sampling is in the works. 

```python
posterior = EmpiricalPosteriorSampling(p_pred_obs, p_pred_prior[itrain:,:], pprior[itrain:,:])
```

```python
utils.plot_posteriors(posterior, prior=pprior, truths=ptrue, varnames=data["varnames"], units=data["units"])
```
